```python
def browse_images(files):
    n = len(files)
    def view_image(i):
        dcm = dicom.read_file(files[i])
        pp.figure(figsize=(8,8), dpi=100)
        pp.imshow(dcm.pixel_array)#, cmap=pp.cm.gray_r, interpolation='nearest')
        pp.title('%d: %s' % (i,files[i]))
        pp.show()
    interact(view_image, i=(0,n-1))
```
